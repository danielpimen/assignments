```python
##Import Pandas & Numpy
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
##Import AAPL 
aapl_df = pd.read_csv('aapl.csv')
aapl_df.head()
```

```python
##Import TSLA 
tsla_df = pd.read_csv('tsla.csv')
tsla_df.head()
```

```python
#1. Please compute the covariance and correlation between the Close prices of APPL and TSLA

##Covarience between AAPL Close & TSLA Clsoe 
print(aapl_df.Close.cov(tsla_df.Close))
```

```python
##Correlation between AAPL Close & TSLA Clsoe 
column_1 = aapl_df['Close']
column_2 = tsla_df['Close']
correlation = column_1.corr(column_2)
print(correlation)
```

```python
## 2. Please compute 5 day moving average on both Close price and volume for AAPL

##Convert Date column to index
aapl_df = aapl_df.set_index('Date', inplace=False)
aapl_df.head()
```

```python
##Drop Additional Columns for cleaner look 
rolling_avg_aapl = aapl_df.loc[:, ('Close', 'Volume')]

#AAPL 5 Day Moving Average for Close Price & Volume 
rolling_avg_aapl = rolling_avg_aapl.rolling(window=5).mean()


print(rolling_avg_aapl)
```

```python
## 3. Visualize the Close prices of both AAPL and TSLA in one chart of lines

##Set TSLA index 
##Convert Date column to index
tsla_df = tsla_df.set_index('Date', inplace=False)
tsla_df.head()
```

```python
##Rename AAPL Columns
aapl_df = aapl_df.rename(columns={'Open': 'AAPL_Open', 'High': 'AAPL_High', 'Low' : 'AAPL_Low', 'Close' : 'AAPL_Close', 'Adj Close' : 'AAPL_Adj', 'Volume': 'AAPL_Volume'})
aapl_df.head()
```

```python
##Rename TSLA Columns
tsla_df = tsla_df.rename(columns={'Open': 'TSLA_Open', 'High': 'TSLA_High', 'Low' : 'TSLA_Low', 'Close' : 'TSLA_Close', 'Adj Close' : 'TSLA_Adj', 'Volume': 'TSLA_Volume'})
tsla_df.head()
```

```python
##Join two dataframes
combine_df = aapl_df.join(tsla_df)
combine_df.head()
```

```python
##Grab Close columns for both TSLA and AAPL 
close_df = combine_df.loc[:, ('AAPL_Close', 'TSLA_Close')]
close_df.head()
```

```python
##Visual for Close prices
close_df.plot()
```
